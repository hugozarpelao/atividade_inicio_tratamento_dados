```python
#importando biblioteca pandas
import pandas as pd
import numpy as np
```

```python
#importando o arquivo csv
df = pd.read_csv('arquivos/topyoutube.csv')
#comando para imprimir tabela
df
```

```python
#info para levantar infos basicas do df
df.info()
```

```python
#transformando em float para multiplicar por 100m
df['Total Views'] = df['Total Views'].str.replace(',', '').astype(float)
```

```python
df
```

```python
#multiplicando coluna Total Views por 100 milhões
df['Total Views'] = df['Total Views'] *100000000
df
```

```python
#verificar aonde possui valor null
df.isnull().sum()
```

```python
#preenchendo colunas com "Não informado"
#poderia ser realizado para todas as colunas,
#porém utlizei por coluna para ser mais objetivo

df['100M'] = df['100M'].fillna('Não informado')
```

```python
#verificando se possui valor null
df.isnull().sum()
```

```python
df
```

```python
#Limitando coluna AVG em apenas 2 casas decimais

df['Avg'] = df['Avg'].round(2)
df
```

```python
#listando os top 10
df.nlargest(11, 'Total Views')
```

```python
# Ordenando o DataFrame pela coluna 'Artist' em ordem decrescente
df_decrescente_artist = df.sort_values(by='Artist', ascending=False)
```
